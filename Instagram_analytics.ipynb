# Instagram Analytics Challenge
Problem Statement:

Assume you are given a challenge to analyze raw Instagram follower data of OpenAI.

Objectives:
- Identify user with maximum posts
- Identify user with maximum followers
- Identify user who follows maximum accounts
- Analyze distribution of account categories



# DATA COLLECTION
The dataset was collected from publicly available Instagram profiles of OpenAI followers.

Steps involved:
- Extract raw profile information
- Parse structured fields (posts, followers, following)
- Convert K/M formatted values into numerical format
- Store cleaned data in structured dictionary format


```python
with open("finaldata.txt", encoding='utf-8') as f:
    data = f.read()
```

```python
chunks = data.split("\n\n") 
chunks = [c for c in chunks if len(c)>3]
```

```python
def parse_chunk(chunk): 
        chunk = chunk.strip()
        sep_chunk = chunk.split('\n')
        username = sep_chunk[0]
        no_of_posts = int(sep_chunk[1].split("post")[0].replace(",",""))
        no_of_followers = float(sep_chunk[2].split("followers")[0].replace(",","").replace("K","").replace("M",""))
        if("K" in sep_chunk[2]):
            no_of_followers=int(no_of_followers*1000)
        elif("M" in sep_chunk[2]):
            no_of_followers=int(no_of_followers*100000) 
        else:
            no_of_followers=int(no_of_followers) 
        no_of_following = sep_chunk[3]
    
        no_of_following = float(sep_chunk[3].split("following")[0].replace(",","").replace("K","").replace("M",""))
        if("K" in sep_chunk[3]):
            no_of_following=int(no_of_followers*1000)
        elif("M" in sep_chunk[3]):
            no_of_following=int(no_of_followers*100000) 
        else:
            no_of_following=int(no_of_following) 
            
        name = sep_chunk[4]
        if(len(sep_chunk)> 5):
            type_of_page = sep_chunk[5]
            bio = "\n".join(sep_chunk[6:])
        else:
            type_of_page = "Unknown"
            bio = ""
        return {"username": username, "no_of_posts": no_of_posts, "no_of_followers": no_of_followers, 
                "no_of_following": no_of_following, "name": name, "type_of_page": type_of_page, "bio": bio}

    

    
```

```python
all_chunks = [] 
for chunk in chunks:  
    parsed_chunk = parse_chunk(chunk)
    all_chunks.append(parsed_chunk)
```

```python
import json
s = json.dumps(all_chunks, indent=4)
with open("data.json","w") as f:
    f.write(s)
```
